```python
#open the text file
text_file = open("../data/nlp_blurb.txt")

#read the data
text = text_file.read()

#datatype of the data read
print(type(text))

#print the text
print(text)

#length of text
print(len(text))
```

```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
```

```python
#import nltk  
#nltk.download()
```

```python
#tokenize the text by sentences
sentences = sent_tokenize(text)

#how many sentences are there
print(len(sentences))

#print the sentences
sentences
```

```python
#tokenize the text by words
words = word_tokenize(text)

#how many sentences are there
print(len(words))

#print the sentences
print(words)
```

```python
#find the frequency distribution
from nltk.probability import FreqDist

#find the frequency
fdist = FreqDist(words)

#print 10 most common words
fdist.most_common(10)
```

```python
#plot the frequency graph
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
#remove puncutation marks
words_no_punc = []

#removing punctuation marks
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
        
#print words without puncutation marks
print(words_no_punc)

print(len(words_no_punc))
```

```python
#find the frequency
fdist = FreqDist(words_no_punc)

#print 10 most common words
print(fdist.most_common(10))

#plot
fdist.plot(10)
```

```python
#list of stopwords
from nltk.corpus import stopwords

#list of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
#removing stopwords
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(clean_words)
print(len(clean_words))
```

```python
#find the frequency for clean words
fdist = FreqDist(clean_words)

#print 10 most common words
print(fdist.most_common(10))

#plot
fdist.plot(10)
```

```python
#PoS tagging
tag = nltk.pos_tag(["Studying","Study"])
print(tag)

sentence = "A very beautiful young lady is walking on the beach"

#tokenizing words:
tokenized_words = word_tokenize(sentence)
print(tokenized_words)

for words in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

print(tagged_words)
```

```python
#Chunking

#extracting noun phrases from text

# ? #optional character
# * - 0 or more repetations

grammer = "NP : {<DT>?<JJ>*<NN>}"

#creating parser
parser = nltk.RegexpParser(grammer)

output = parser.parse(tagged_words)
print(output)

#to visualize
output.draw()
```

```python
#chinking
#excludes a part of the chunk

grammer = r""" NP: {<.*>+} \n
               }<JJ>+{"""

#creating parser
parser = nltk.RegexpParser(grammer)

output = parser.parse(tagged_words)
print(output)

#to visualize
output.draw()
```

```python
#Named entity recognition
#When the binary value is True, then it will only show 
#whether a particular entity is named entity or not. 
#It will not show any further details on it.

sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#tokenizing words
tokenized_words = word_tokenize(sentence)

#PoS tagging
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)
    
print(tagged_words)

#Named entity recognition
N_E_R = nltk.ne_chunk(tagged_words, binary = True)

print(N_E_R)

#visualize
N_E_R.draw()
```

```python
#Named entity recognition
#When the binary value is True, then it will only show 
#whether a particular entity is named entity or not. 
#It will not show any further details on it.

sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#tokenizing words
tokenized_words = word_tokenize(sentence)

#PoS tagging
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)
    
print(tagged_words)

#Named entity recognition
N_E_R = nltk.ne_chunk(tagged_words, binary = False)

print(N_E_R)

#visualize
N_E_R.draw()
```

```python
#WordNet
from nltk.corpus import wordnet

for words in wordnet.synsets("Fun"):
    print(words)
```

```python
#different meaning
for words in wordnet.synsets("Fun"):
    for lemma in words.lemmas():
        print(lemma)
```

```python
#all details for a word
word= wordnet.synsets("play")[0]

#checking name
print(word.name())

#checking definition
print(word.definition())

#checking examples
print(word.examples())
```

```python
#all details for all meanings of a word
for words in wordnet.synsets("Fun"):
    print(words.name())
    print(words.definition())
    print(words.examples())
    
    for lemma in words.lemmas():
        print(lemma)
```

```python
#hypernyms abstract term for the word
#hyponames more specific term for the word
#get the name only

word = wordnet.synsets("play")[0]

print(word.hypernyms())
print(word.hyponyms())
print(word.lemmas()[0].name())
```

```python
#synonymes and antonyms

synonyms = []

for words in wordnet.synsets('Fun'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

print(synonyms)

antonyms = []
for words in wordnet.synsets('Natural'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
            
print(antonyms)
```

```python
#finding similarity between words
word1 = wordnet.synsets("ship","n")[0]
word2 = wordnet.synsets("boat","n")[0]

#check similarity
print(word1.wup_similarity(word2))
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]

cv = CountVectorizer()

#generating output for baf of words
B_O_W = cv.fit_transform(sentences).toarray()

#Total words with their index in model
print(cv.vocabulary_)

print(cv.get_feature_names())

print(B_O_W)
```

```python
#TF_IDF
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = ['This is the first document',"This document is the second document"]

vectorizer = TfidfVectorizer(norm = None)

X = vectorizer.fit_transform(sentences).toarray()

print(vectorizer.vocabulary_)

print(vectorizer.get_feature_names())

print(X)
```
